```python
# Ecommerce Sales Analysis
### Data Analysis using Pandas GroupBy

This notebook analyzes synthetic e-commerce sales data to identify sales performance across different regions, product categories, and salespersons.

The analysis includes:
- Basic grouping and aggregations
- Multi-column grouping
- Multiple aggregation metrics
- Custom aggregation functions
- Final sales summary report
```

```python
import pandas as pd
import random
```

```python
## Dataset Generation

Since we cannot access the company database directly, we generate a synthetic dataset with 200 sales transactions including regions, product categories, salespersons, and sales amounts.
```

```python
random.seed(42)

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

df.head(10)
```

```python
print("Dataset Shape:", df.shape)
```

```python
## 1. Total Sales Amount for Each Region
```

```python
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()

region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)

region_sales_sorted
```

```python
This calculates the total sales generated in each region using groupby() and sum().
The results are sorted in descending order to identify the highest performing region.
```

```python
## 2. Number of Transactions per Category
```

```python
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

category_transactions
```

```python
This counts how many sales transactions occurred in each product category.
```

```python
## 3. Average Sales Amount per Salesperson
```

```python
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

salesperson_avg
```

```python
This shows the average transaction value handled by each salesperson.
```

```python
# Task 2: Multi-Column Grouping and Multiple Aggregations
```

```python
## 1. Total Sales by Region and Category
```

```python
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

region_category_sales
```

```python
This analysis groups the data by both region and category to calculate total revenue for each combination.
```

```python
## 2. Multiple Metrics per Salesperson
```

```python
salesperson_stats = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

salesperson_stats_sorted = salesperson_stats.sort_values(by='sum', ascending=False)

salesperson_stats_sorted
```

```python
Using agg(), we calculate:
- Total sales
- Average sales
- Number of transactions

Sorting by total sales helps identify the top-performing salesperson.
```

```python
## 3. Category with Maximum Revenue
```

```python
category_sales = df.groupby('category')['sales_amount'].sum()

top_category = category_sales.idxmax()

print("Top Revenue Category:", top_category)
```

```python
idxmax() returns the category with the highest total sales revenue.
```

```python
# Task 3: Custom Aggregation and Complete Sales Report
```

```python
## Custom Aggregation Function
```

```python
def sales_range(x):
    return x.max() - x.min()
```

```python
This function calculates the difference between the maximum and minimum sales values in each group.
```

```python
region_analysis = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max',sales_range]).reset_index()

region_analysis
```

```python
This provides detailed statistics about sales performance in each region.
```

```python
## Final Sales Summary Report
```

```python
final_report = df.groupby('region').agg({
    'sales_amount': ['sum','mean'],
    'customer_id': 'count'
}).reset_index()

final_report
```

```python
This final report shows:
- Total sales revenue
- Average transaction value
- Total number of transactions per region
```

```python
# Explanation of Multi-Level Columns

When dictionary-based aggregation is used, Pandas creates hierarchical column labels.

Example:

sales_amount → sum, mean
customer_id → count

This results in a multi-level column structure where:

Level 1 → Original column name
Level 2 → Aggregation function

This differs from single aggregation where only one output column is produced.
```
